```python
import pandas as pd
import numpy as np
df = pd.read_csv("salaries.csv")
```

```python
df = df.fillna(0)
df
```

```python
df["number"] = df["Annual Salary"].apply(lambda x: x if x == 0 else float(str(x)[1:]))
```

```python
df["number"].apply(np.sqrt)
```

```python
df.number.transform(np.sqrt)
```

```python
df.number.groupby(df["number"]).apply(np.sqrt)
```
