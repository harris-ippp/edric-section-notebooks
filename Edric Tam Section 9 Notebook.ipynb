```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from rpy2.robjects import r, pandas2ri
pandas2ri.activate()
df = r['iris']
```

```python
df
```

```python
## use default pandas df
df.boxplot()
```

```python
sns.boxplot(x = 'Species', y = 'Sepal.Length', data=df)
```

```python
sns.pairplot(df, vars = ['Petal.Width', 'Petal.Length'])
```

```python
sns.regplot('Petal.Width', 'Sepal.Length', df)
```

```python
sns.pairplot(df, vars = ['Petal.Width', 'Sepal.Length', 'Sepal.Width'], kind = 'reg', hue = 'Species')
```

```python
# https://github.com/statsmodels/statsmodels/
from statsmodels.regression.linear_model import OLS
```

```python
df[['Petal.Width', 'Petal.Length']]
```

```python
regressionModel = OLS(df['Petal.Width'], df[['Sepal.Length', 'Sepal.Width']])
```

```python
modelFit = regressionModel.fit()
```

```python
modelFit.summary()
```

```python
import statsmodels.formula.api as smf
df.columns = ['SL', 'SW', 'PL', 'PW', 'Species']
df
```

```python
formulaFit = smf.ols('SL ~ PW*PL', df).fit()
formulaFit.summary()
```

```python
## Testing

## Feed in any two series and a test statistic and p value will be generated
from scipy.stats import ttest_ind
ttest_ind(df.SW, df.SL)
```
